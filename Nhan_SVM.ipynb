<strong><p><H2>PreProcessing</H2></p></strong>

```python
train = np.vstack((queen_train, no_queen_train, queen_val,no_queen_val))
train = pd.DataFrame(train)
train = train.sample(frac=1, random_state=42)
```

```python
test = np.vstack((queen_test, no_queen_test))
test = pd.DataFrame(test)
test = test.sample(frac=1, random_state=42)
```

```python
train
```

<strong><p><H2>Modeling</H2></p></strong>

<strong><p><H4>Split Training set and Testing set</H4></p></strong>

```python
Y_train = train[1131]
X_train = train.drop([1131], axis=1)
Y_test = test[1131]
X_test = test.drop([1131], axis=1)
```

<strong><p><H4>Data Normalization</H4></p></strong>

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_train = pd.DataFrame(X_train)
X_test = scaler.fit_transform(X_test)
X_test = pd.DataFrame(X_test)
```

<strong><p><H4>Modeling</H4></p></strong>

```python
from sklearn.svm import SVC
SVM_model = SVC()
```

```python
from sklearn.model_selection import GridSearchCV   
# defining parameter range 
param_grid1 = {'C': [0.1, 1, 10, 100],  
              'gamma': [1, 0.1, 0.01, 0.001], 
              'kernel': ['rbf']}  
SVM_grid1 = GridSearchCV(SVM_model, param_grid1, refit = True, verbose = 3) 
```

```python
SVM_modelc100 = SVC(C = 100, gamma =  0.001, kernel = 'rbf')
```

```python
param_grid2 = {'C': [0.1, 1, 10, 100],
               'gamma': [1, 0.1, 0.01],
               'degree': [3,4,5],
               'kernel': ['poly']}
SVM_grid2 = GridSearchCV(SVM_model, param_grid2, refit = True, verbose = 3, cv=3) 
```

<strong><p><H4>Fit</H4></p></strong>

```python
SVM_grid1.fit(X_train,Y_train)
```

```python
print('Best Grid:',SVM_grid1.best_params_) 
print(SVM_grid1.best_estimator_) 
```

```python
SVM_modelc100.fit(X_train,Y_train)
```

```python
SVM_grid2.fit(X_train,Y_train)
```

```python
print(SVM_grid2.best_params_) 
print(SVM_grid2.best_estimator_) 
```

<strong><p><H2>Test Accuracy</H2></p></strong>

<strong><p><H4>SVM model with {'C': 10, 'gamma': 0.001, 'kernel': 'rbf'}</H4></p></strong>

```python
accuracy_score(SVM_grid1.predict(X_test),Y_test)
```

<strong><p><H4>SVM model with {'C': 100, 'degree': 4, 'gamma': 0.001, 'kernel': 'rbf'}</H4></p></strong>

```python
accuracy_score(SVM_modelc100.predict(X_test),Y_test)
```

```python
accuracy_score(SVM_grid2.predict(X_test),Y_test)
```
