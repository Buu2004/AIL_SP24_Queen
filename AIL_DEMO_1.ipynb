```python
import pandas as pd
import librosa
import matplotlib.pyplot as plt
import numpy as np
from google.colab import drive
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import GridSearchCV
```

```python
drive.mount('/content/drive')
```

```python
os.chdir('/content/drive/My Drive/Colab Notebooks/AIL/20k_audio_splitted_dataset/train/Queen_train')
files = os.listdir()
queen_train = []
for file in files:
  audio_file = '/content/drive/My Drive/Colab Notebooks/AIL/20k_audio_splitted_dataset/train/Queen_train/'+file
  signal, sr = librosa.load(audio_file)
  mfccs = librosa.feature.mfcc(y=signal, n_mfcc = 13, sr=sr)
  queen_train.append(mfccs)
```

```python
os.chdir('/content/drive/My Drive/Colab Notebooks/AIL/20k_audio_splitted_dataset/val/Queen')
files = os.listdir()
for file in files:
  audio_file = '/content/drive/My Drive/Colab Notebooks/AIL/20k_audio_splitted_dataset/val/Queen/'+file
  signal, sr = librosa.load(audio_file)
  mfccs = librosa.feature.mfcc(y=signal, n_mfcc = 13, sr=sr)
  queen_train.append(mfccs)

queen_train = np.array(queen_train)
queen_train = queen_train.reshape(queen_train.shape[0], -1)
labels = np.ones((queen_train.shape[0], 1))
queen_train = np.hstack((queen_train, labels))
```

```python
os.chdir('/content/drive/My Drive/Colab Notebooks/AIL/20k_audio_splitted_dataset/test/Queen')
files = os.listdir()
queen_test = []
for file in files:
  audio_file = '/content/drive/My Drive/Colab Notebooks/AIL/20k_audio_splitted_dataset/test/Queen/'+file
  signal, sr = librosa.load(audio_file)
  mfccs = librosa.feature.mfcc(y=signal, n_mfcc = 13, sr=sr)
  queen_test.append(mfccs)

queen_test = np.array(queen_test)
queen_test = queen_test.reshape(queen_test.shape[0], -1)
labels = np.ones((queen_test.shape[0], 1))
queen_test = np.hstack((queen_test, labels))
```

```python
os.chdir('/content/drive/My Drive/Colab Notebooks/AIL/20k_audio_splitted_dataset/train/NonQueen_train')
files = os.listdir()
no_queen_train = []
for file in files:
  audio_file = '/content/drive/My Drive/Colab Notebooks/AIL/20k_audio_splitted_dataset/train/NonQueen_train/'+file
  signal, sr = librosa.load(audio_file)
  mfccs = librosa.feature.mfcc(y=signal, n_mfcc = 13, sr=sr)
  no_queen_train.append(mfccs)
```

```python
os.chdir('/content/drive/My Drive/Colab Notebooks/AIL/20k_audio_splitted_dataset/val/NonQueen')
files = os.listdir()
for file in files:
  audio_file = '/content/drive/My Drive/Colab Notebooks/AIL/20k_audio_splitted_dataset/val/NonQueen/'+file
  signal, sr = librosa.load(audio_file)
  mfccs = librosa.feature.mfcc(y=signal, n_mfcc = 13, sr=sr)
  no_queen_train.append(mfccs)

no_queen_train = np.array(no_queen_train)
no_queen_train = no_queen_train.reshape(no_queen_train.shape[0], -1)
labels = np.zeros((no_queen_train.shape[0], 1))
no_queen_train = np.hstack((no_queen_train, labels))
```

```python
os.chdir('/content/drive/My Drive/Colab Notebooks/AIL/20k_audio_splitted_dataset/test/NonQueen')
files = os.listdir()
no_queen_test = []
for file in files:
  audio_file = '/content/drive/My Drive/Colab Notebooks/AIL/20k_audio_splitted_dataset/test/NonQueen/'+file
  signal, sr = librosa.load(audio_file)
  mfccs = librosa.feature.mfcc(y=signal, n_mfcc = 13, sr=sr)
  no_queen_test.append(mfccs)

no_queen_test = np.array(no_queen_test)
no_queen_test = no_queen_test.reshape(no_queen_test.shape[0], -1)
labels = np.zeros((no_queen_test.shape[0], 1))
no_queen_test = np.hstack((no_queen_test, labels))
```

```python
train_set = np.vstack((queen_train, no_queen_train))
train_set = pd.DataFrame(train_set)
train_set = train_set.sample(frac=1, random_state=42)
```

```python
test_set = np.vstack((queen_test, no_queen_test))
test_set = pd.DataFrame(test_set)
test_set = test_set.sample(frac=1, random_state=42)
```

```python
train_set
```

```python
test_set
```

```python
y_train = train_set[1131]
X_train = train_set.drop([1131], axis=1)
y_test = test_set[1131]
X_test = test_set.drop([1131], axis=1)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_train = pd.DataFrame(X_train)
X_test = scaler.fit_transform(X_test)
X_test = pd.DataFrame(X_test)
```

```python
X_train
```

```python
X_test
```

```python
# n_estimators = [10, 50, 100]
# max_features = ['sqrt', 'log2']
# max_depth = [2, 4]
# min_samples_split = [2, 5]
# min_samples_leaf = [1, 2]
# bootstrap = [True, False]
n_estimators = [100, 500, 1000]
max_features = ['sqrt']
max_depth = [100, 500, 1000]
min_samples_split = [2]
min_samples_leaf = [1]
bootstrap = [True]
```

```python
param_grid = {
    'n_estimators': n_estimators,
    'max_features' : max_features,
    'max_depth' : max_depth,
    'min_samples_split' : min_samples_split,
    'min_samples_leaf' : min_samples_leaf,
    'bootstrap' : bootstrap
}
```

```python
RF_model = RandomForestClassifier()
```

```python
RF_Grid = GridSearchCV(estimator=RF_model, param_grid=param_grid, cv=10)
```

```python
RF_Grid.fit(X_train, y_train)
```

```python
RF_Grid.best_params_
```

```python
print(f'Train Accuracy: {RF_Grid.score(X_train, y_train)}')
print(f'Test Accuracy: {RF_Grid.score(X_test, y_test)}')
```

```python
RF_model = RandomForestClassifier(bootstrap= True,
 max_depth= 400,
 max_features= 'sqrt',
 min_samples_leaf= 1,
 min_samples_split= 2,
 n_estimators= 100)
RF_model.fit(X_train, y_train)
```

```python
RF_model.score(X_test, y_test)
```

```python
RF_Grid = GridSearchCV(estimator=RF_model, param_grid=param_grid, cv=3, verbose = 3)
RF_Grid.fit(X_train, y_train)
```

```python
RF_Grid.best_params_
```

```python
print(f'Train Accuracy: {RF_Grid.score(X_train, y_train)}')
print(f'Test Accuracy: {RF_Grid.score(X_test, y_test)}')
```

```python
RF_model = RandomForestClassifier(bootstrap= True,
 max_depth= 500,
 max_features= 'sqrt',
 min_samples_leaf= 1,
 min_samples_split= 2,
 n_estimators= 1000)
RF_model.fit(X_train, y_train)
print(RF_model.score(X_test, y_test))
```

```python
ET_model = ExtraTreesClassifier()
ET_Grid = GridSearchCV(estimator=ET_model, param_grid=param_grid, cv=3, verbose = 3)
ET_Grid.fit(X_train, y_train)
```

```python
ET_Grid.best_params_
```
